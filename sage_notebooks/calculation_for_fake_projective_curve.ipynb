# elliptic curve with O not at infinity
This is to change coordinates so that the curve y^2 = x^3-4x+1 has its O at the finite point (0,y_oo)

```python
# (0,y_\infty) is the point where O will map to
yoo = 100
M = matrix([[yoo-4,0,0],[0,yoo,0],[0,1,yoo-4]])
(M/(yoo-4)).inverse()
```

```python
# new equation for our favourite curve
A.<x,y> = QQ[]

xx = yoo*x/(yoo-y)
yy = (yoo-4)*y/(yoo-y)
f = (-yy^2 + xx^3-4*xx+1)*(yoo-y)^3
f = A(f)
f
```

```python
def cubic_interpolation(a,fa,ffa,b,fb,ffb):
    r"""
    return the coefficience in f=a0*t^3+..+a4
    such that f(a)=fa, f'(a) =ffa, f(b)=fb,f'(b)=ffb
    """
    M = matrix([
        [a^3,a^2,a,1],
        [3*a^2,2*a,1,0],
        [b^3,b^2,b,1],
        [3*b^2,2*b,1,0]])
    return M.inverse() * vector([fa,ffa,fb,ffb])
```

```python
cubic_interpolation(0,0,0,.4,0.3,3)
```

```python
cubic_interpolation(.4,.3,3,.5,.6,0)
```

```python
cubic_interpolation(.5,.6,0,.7,.4,0)
```
